```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(df['TotalCharges'].isnull().sum())
```

```python
df.dropna(subset=['TotalCharges'], inplace=True)
```

```python
df
```

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['Partner'] = le.fit_transform(df['Partner'])
df['Dependents'] = le.fit_transform(df['Dependents'])
df['PhoneService'] = le.fit_transform(df['PhoneService'])
df["PaperlessBilling"] = le.fit_transform(df["PaperlessBilling"])
df['Churn'] = le.fit_transform(df['Churn'])
```

```python
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[['InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies', 'PaymentMethod', 'Contract']])
```

```python
encoded
```

```python
df.drop(columns=['InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies', 'PaymentMethod', 'Contract'], inplace=True)
```

```python
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(
    ['InternetService','OnlineSecurity','OnlineBackup','DeviceProtection',
     'TechSupport','StreamingTV','StreamingMovies','PaymentMethod','Contract']
))
```

```python
df.reset_index(drop=True, inplace=True)
encoded_df.reset_index(drop=True, inplace=True)
```

```python
encoded_df
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.to_csv("Customer_Churn_Data_V2.csv", index=False, sep=",")
```
